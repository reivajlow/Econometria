## Revision de literatura

#### Que es el tipo de cambio?
El tipo de cambio es la relacion en equivalencia entre dos monedas diferentes, Lo cual sirve de referencia en transacciones comerciales, esto puesto que un objeto de un valor especifico, tiene valores distintos segun la preciacion de la moneda, es decir, son distintas cantidades de moneda con el mismo poder adquisitivo, es por esto la importancia del tipo de cambio y el porque se quiere investigar para determinar variables que afecten su variacion y conocer tecnicas o formas de controlar la variacion del tipo de cambio para mantener una economia estable.

#### Metodologia

En la metodologia utilizada por distintos papers se pueden definir metodos comunes asi como variables de interes que determinan como varia el tipo de cambio segun indices y ocurrecias tanto economicas como sociales.

entre las variables mas mencionadas encontramos a grandes razgos son la inflacion, productividad, la posición fiscal del país, el nivel de endeudamiento público y las expectativas de crecimiento económico. Todas estas variables estudiadas en base a indices economicos y sociales que seran profundizados a la brevedad.

Un tema clave en el estudio del tipo de cambio, es la teoria de la paridad del poder de compra, la cual explicada de manera resumida consiste en; ser capaces de adquirir una misma canasta de bienes en cualquier país por la misma cantidad de divisas.
Esto, en la practica se hace imposible esta paridad debido a que ella no considera factores como: barreras al comercio internacional(impuestos,aduanas), flujos de capital y la economias no funcionan a pleno empleo. es por esto que nace el concepto del tipo de cambio real y tipo de cambio nominal.


Tipo de cambio nominal:
    El tipo de cambio nominal es la cotización de una divisa(moneda no local) frente a otra en los mercados financieros.
tipo de cambio real:
    Es el poder adquisitivo de la moneda local en el extranjero.
    
Intuitivamente podemos entender que el tipo de cambio que nos interesa estudiar es el real,dado que, este es el que mejor representa el impacto de su variacion en la poblacion, por lo cual se hace mas sencillo de interpretar los efectos en la poblacion. Si quisiéramos comparar precios entre países para ver cuál es más caro poco nos diría el tipo de cambio nominal. Por ejemplo, el tipo de cambio nominal en Colombia es de 3.000 pesos colombianos mientras que en Argentina es de 37 pesos argentinos. nada nos dice esto sobre que pais es mas caro, para eso debemos analizar los precios vigentes en cada pais y calcular el tipo de cambio real entre ambos paises, asi basado en alguno producto, servicio o canaste de productos, podemos denominar que pais es mas caro o economico.

Los economistas han debatido ampliamente los pros y los contras de establecer una tasa real alta para el desarrollo económico. A grandes rasgos, por un lado, se argumenta que esta política incrementará la producción y el empleo al aumentar las exportaciones y sustituir importaciones. Por otro lado, estos efectos potenciales sobre la producción están matizados por efectos distributivos a nivel local y enfatizan la naturaleza decreciente de la distribución del ingreso, lo cual afecta directamente a la poblacion.

expresada en logaritmos. TCRE es el índice del tipo de cambio real efectivo, que se define como el ratio
entre los precios domésticos y los precios en el exterior, ponderado por el comercio entre países y ajustada
por el tipo de cambio; dada esta definición, un aumento en el TCRE equivale a una apreciación. Para el
ejercicio de robustez, también se utiliza el índice de tipo de cambio real bilateral (TCR) frente al dólar de
